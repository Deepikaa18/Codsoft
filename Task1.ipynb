**Titanic Dataset**

```python
import pandas as pd
a = pd.read_csv("tested.csv")
print(a)
```

```python
a.isna()
```

```python
from sklearn import preprocessing
d = preprocessing.LabelEncoder()
a["Name"] = d.fit_transform(a["Name"])
a["Sex"] = d.fit_transform(a["Sex"])
a["Ticket"] = d.fit_transform(a["Ticket"])
a["Cabin"] = d.fit_transform(a["Cabin"])
a["Embarked"] = d.fit_transform(a["Embarked"])
a["Name"].unique
a["Sex"].unique
a["Ticket"].unique
a["Cabin"].unique
a["Embarked"].unique
```

 **Visulaization**

```python
import seaborn as sns
sns.barplot(x='Embarked', y='Name', data=a)
```
