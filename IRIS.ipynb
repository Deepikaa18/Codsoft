```python
print( "               ******          DATASET FOR IRIS          *******")
print("                ")
print("                ")
import pandas as pd
a = pd.read_csv("IRIS.csv")
print(a)
```

```python
a.isna()
```

```python
print("                                   *-*   PREPROCESSING   *-*                        ")
print("            ")
from sklearn import preprocessing
d = preprocessing.LabelEncoder()
a["species"] = d.fit_transform(a["species"])
a["species"].unique
```
